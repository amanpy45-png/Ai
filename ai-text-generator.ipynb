```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="gpt2"
)

result = generator("Once upon a time")
print(result)
```

```python
# we only write model = 'gpt2', we never downloaded gpt2 manually.
# bts

# model="gpt2"

# ↓

# Transformers Library

# ↓

# Looks on your PC

# ↓

# Model found?

#       ↓

# NO

# ↓

# Download from Hugging Face Hub

# ↓

# Store locally

# ↓

# Load into RAM

# ↓

# Ready
```

```python
# gpt doesn't pick completely randomly, it predicts which next word has the highest prob.
```

```python
result = generator(
    'once upon a time',
    max_length= 100, 
    num_return_sequences=5
)

for i in result:
    print(i['generated_text'])
    print('-'*100)
```

```python
# observation:
# inc max_length -> it writes a longer story
# story is not complete, as gpt doesn't know when your story is finished.
# gpt-2 : sentences repeating. cause it's predicting most probable next word.

# techniques like temperature and sampline reduces it.
```

```python
result = generator(
    'Write a professional leave email to my manager because I have fever.',
    max_length= 100, 
    num_return_sequences=3
)

for i in result:
    print(i['generated_text'])
    print('-'*100)
```
