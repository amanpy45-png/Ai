```python
# handling chat conversion 

chat_words = {
    'ASAP': 'As Soon As Possible',
    'AFK': 'Away From Keyboard',
    'BTW': 'By The Way',
    'FYI': 'For Your Information',
    'IMO': 'In My Opinion',
    'NVM': 'Never Mind',
    'TBH': 'To Be Honest',
    'LOL': 'Laughing Out Loud',
    'IDK': 'I Don\'t Know',
    'SMH': 'Shaking My Head',
    'FR': 'For Real',
    'RN': 'Right Now',
    'BRB': 'Be Right Back',
    'TTYL': 'Talk To You Later',
    'ICYMI': 'In Case You Missed It',
    'TLDR': 'Too Long; Didn\'t Read',
    'TIA': 'Thanks In Advance',
    'JK': 'Just Kidding',
    'JIC': 'Just In Case',
    'FOMO': 'Fear Of Missing Out',
    'GG': 'Good Game',
    'GLHF': 'Good Luck Have Fun',
    'IRL': 'In Real Life',
    'LMK': 'Let Me Know',
    'MYOB': 'Mind Your Own Business',
    'OMG': 'Oh My God',
    'OTW': 'On The Way',
    'POV': 'Point Of View',
    'TMI': 'Too Much Information',
    'YW': 'You\'re Welcome'
}
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
chat_conversion('Do this work ASAP')
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)

    x= new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
# handle emoji

import re

def remove_emoji(text):
    emoji_pattern = re.compile('[\U0001F600\U0001F44D\U0001F680]')
    return emoji_pattern.sub(r'', text)
```

```python
print(remove_emoji('Loved this movie. It was 👍 👍'))
```

# Stemmer

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = 'play played playing'
stem_words(sample)        # convert into root word, and sometimes it's not readable. lemmatization can handle this.
```

# lemmatization

```python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')

wordnet_lemmatizer = WordNetLemmatizer()

sentence = 'He was running and eating at the same time. He has bad habit of swimming after playing long hours in the ground'
punctuations = "?:!.,;"

# Tokenize
sent_words = nltk.word_tokenize(sentence)

filtered_words = [word for word in sent_words if word not in punctuations]

# Print header
print("{0:20}{1:20}".format('Word', 'Lemma'))

# Print results
for word in filtered_words:
    print("{0:20}{1:20}".format(word, wordnet_lemmatizer.lemmatize(word, pos='v')))
```

> Note : Stemming and Lemmatization are same to retrieve root words but lemmatization is worked good, lemmatization is slow and stemming is fast
